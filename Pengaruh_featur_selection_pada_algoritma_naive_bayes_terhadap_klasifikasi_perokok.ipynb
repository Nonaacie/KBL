## forward selection

Forward Selection adalah metode dalam analisis regresi di mana model dibangun secara iteratif dengan menambahkan satu per satu variabel independen (fitur) ke dalam model. Langkah-langkah umum forward selection adalah sebagai berikut:

Langkah 1: Dimulai dengan model yang tidak memiliki variabel independen (model kosong).

Langkah 2: Evaluasi setiap variabel independen yang belum dimasukkan ke dalam model untuk menentukan variabel yang paling signifikan. Kriteria evaluasi dapat berupa nilai p-nilai atau statistik lain yang relevan.

Langkah 3: Pilih variabel yang paling signifikan dan tambahkan ke dalam model.

Langkah 4: Ulangi langkah 2 dan 3, mengevaluasi variabel yang belum dimasukkan ke dalam model dan menambahkan variabel yang paling signifikan pada setiap iterasi.

Langkah 5: Teruskan proses ini hingga kriteria penghentian terpenuhi. Misalnya, kita dapat berhenti jika penambahan variabel baru tidak memberikan peningkatan signifikan dalam penjelasan variasi.

Tujuan dari forward selection adalah untuk membangun model yang optimal secara bertahap dengan mempertimbangkan variabel satu per satu dan hanya menambahkan variabel yang secara signifikan meningkatkan kemampuan model untuk menjelaskan data.

```python
import statsmodels.api as sm
```

```python
def forward_selection(data, target, significance_level=0.05): #Choose a significance level (e.g. SL = 0.05 with a 95% confidence).
    initial_features = data.columns.tolist() #convert kolom ke dalam bentuk list
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

## backward_elimination

Backward Elimination adalah sebuah metode dalam konteks pemodelan statistik, terutama dalam analisis regresi, di mana kita mulai dengan sebuah model yang mencakup semua fitur yang tersedia dan secara iteratif menghapus satu per satu fitur yang dianggap kurang penting atau tidak signifikan. Tujuan akhir dari backward elimination adalah memperoleh model regresi yang lebih sederhana, tetapi masih mampu menjelaskan variasi yang ada dalam data.

Langkah-langkah umum dalam backward elimination adalah sebagai berikut:

Langkah 1: Mulailah dengan model yang mencakup semua fitur yang tersedia.

Langkah 2: Evaluasi signifikansi statistik dari masing-masing fitur dalam model menggunakan suatu kriteria, seperti nilai p-nilai.

Langkah 3: Hapus fitur yang dianggap kurang signifikan (biasanya yang memiliki nilai p-nilai yang tinggi) dari model.

Langkah 4: Kembali ke Langkah 2 dan ulangi proses tersebut dengan model yang baru (tanpa fitur yang dihapus).

Langkah 5: Teruskan proses eliminasi sampai kriteria penghentian terpenuhi. Misalnya, kita dapat berhenti jika semua fitur dalam model dianggap signifikan berdasarkan nilai p-nilai yang ditentukan.


```python
def backward_elimination(data, target, significance_level=0.05):
    features = data.columns.tolist()

    while len(features) > 0:
        model = sm.OLS(target, sm.add_constant(data[features])).fit()
        max_p_value = model.pvalues.drop('const').max()

        if max_p_value > significance_level:
            removed_feature = model.pvalues.drop('const').idxmax()
            features.remove(removed_feature)
        else:
            break

    return features
```

## Pengetesan algoritma naive bayes


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class NaiveBayesClassifier:
    def __init__(self):
        self.class_probs = {}
        self.feature_probs = {}

    def fit(self, X, y):
        print("Menghitung probabilitas kelas...")
        total_data = len(y)
        unique_classes, class_counts = np.unique(y, return_counts=True)
        for c, count in zip(unique_classes, class_counts):
            prob = count / total_data
            self.class_probs[c] = prob
            print(f"Probabilitas kelas {c}: {prob:.4f}")

        print("\nMenghitung probabilitas fitur diberikan kelas...")
        for c in unique_classes:
            c_mask = (y == c)
            for feature in X.columns:
                feature_counts = X[c_mask][feature].value_counts()
                prob_feature_given_c = feature_counts / len(X[c_mask])
                self.feature_probs[(c, feature)] = prob_feature_given_c
                print(f"\nProbabilitas fitur {feature} diberikan kelas {c}:")
                print(prob_feature_given_c)

        print("\nProses fitting selesai.")

    def predict(self, x):
        print("\nMenghitung probabilitas prediksi...")
        probs = {}
        for c in self.class_probs.keys():
            prob_c = np.log(self.class_probs[c])
            print(f"\nProbabilitas kelas {c} sebelum faktor normalisasi: {prob_c:.4f}")
            for feature in x.index:
                value = x[feature]
                prob_feature_given_c = self.feature_probs.get((c, feature), {}).get(value, 1e-10)
                prob_c += np.log(prob_feature_given_c)
                print(f"Probabilitas fitur {feature} diberikan kelas {c}: {prob_feature_given_c:.4f}")

            probs[c] = prob_c

        print("\nProses prediksi selesai.")
        predicted_class = max(probs, key=probs.get)
        print("\nProbabilitas kelas prediksi setelah faktor normalisasi:")
        for c in self.class_probs.keys():
            normalized_prob = np.exp(probs[c] - max(probs.values()))
            print(f"Kelas {c}: {normalized_prob:.4f}")

        return predicted_class

F = df[['gender']]
y = df["smoking"]
# Contoh penggunaan
X_train, X_test, y_train, y_test = train_test_split(F, y,
                                                    test_size=0.2)

# Membuat dan melatih model Naive Bayes
nb_model = NaiveBayesClassifier()
nb_model.fit(X_train, y_train)

# Mengambil satu sampel dari data uji
sampel_uji111 = X_test.sample(n=1)

# Melakukan prediksi pada satu sampel
predicted_class2 = nb_model.predict(sampel_uji111.iloc[0])

# Menampilkan hasil prediksi
print("Data Uji:")
print(sampel_uji111)
print("\nHasil Prediksi:")
print(predicted_class2)
print("Label Sebenarnya:")
print(y_test[sampel_uji111.index])
```
